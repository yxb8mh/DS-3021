```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('ObesityDataSet-1.csv')

df.info()
df.describe()
```

```python
# Collapse obesity levels
collapse_map = {
    'Insufficient_Weight': 'Insufficient Weight',
    'Normal_Weight':      'Normal Weight',
    'Overweight_Level_I':  'Overweight',
    'Overweight_Level_II': 'Overweight',
    'Obesity_Type_I':      'Obese',
    'Obesity_Type_II':     'Obese',
    'Obesity_Type_III':    'Obese'
}
df['NObeyesdad'] = df['NObeyesdad'].map(collapse_map)

# Drop height and weight columns
df.drop(['Height', 'Weight'], axis=1, inplace=True)

# Quick summary
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())

# Plot class distribution
plt.figure(figsize=(8, 4))
sns.countplot(x='NObeyesdad', data=df)
plt.title('Distribution of Obesity Levels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='NObeyesdad', y='Age', data=df)
plt.title('Age Distribution by Obesity Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='NObeyesdad', y='FAF', data=df)
plt.title('Physical Activity (FAF) across Obesity Levels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='family_history_with_overweight', hue='NObeyesdad', data=df)
plt.title('Family History vs Obesity Level')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue='NObeyesdad', vars=['Age', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE'])
plt.show()
```
